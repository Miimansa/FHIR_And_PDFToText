```python
credentials = Credentials.service_principal_credentials_builder(). \
        with_client_id("your_client_id"). \
        with_client_secret("your_client_secre"). \
        build()

# Create an ExecutionContext using credentials and create a new operation instance.
execution_context = ExecutionContext.create(credentials)
extract_pdf_operation = ExtractPDFOperation.create_new()
```

```python
#Function to convert to text files
#rerun from here
import json
import re


def process_json_and_save_files(json_file_path, output_directory):
    with open(json_file_path, 'rb') as file:
        data = json.load(file)

    data=data['elements']
    content_accumulator = []
    page_counter = 0
    ignore_paths = ["Figure", "Footnote","Table","Reference"]
    ignore_subsections = ["Authors’ Disclosures", "Authors’ Contributions", "Acknowledgments"]
    ignore_content = False
    data_length = len(data)
    i=0
    
    pattern_ = r'/Table(?:\[\d+\])?.*'
    pattern = r'\/\/Document(?:\/\w+(?:\[\d+\])?|\w*\[\d+\]|\w*\*)*\/Table(?:\[\d+\]|[*])?$'
    def save_file(texts, counter):
        filename = f"{output_directory}/section_{counter}.txt"
        with open(filename, 'w') as f:
            header = texts[0] if texts else ""
            content = " ".join(texts[1:]) if len(texts) > 1 else ""
            f.write(f"{header}\n{content}")

    found_table=False
    for item in data:
        i=i+1
        path = item['Path']
        if re.search(pattern_, path):
          if  re.search(pattern, path):
              print(path)
              print(page_counter,json_file_path)
              content_accumulator.append('\n###TABLE###\n')
              found_table=True
          else:
           pass
        else:      
         found_table=False      
        text = item.get('Text', '').strip()

        if any(ignore in path for ignore in ignore_paths):
            if(i==data_length and len(content_accumulator)!=1):
             save_file(content_accumulator, page_counter)
             page_counter += 1
             content_accumulator = []  
            else:
             continue

        if "//Document/Title" in path or (ignore_content and re.match(r"//Document/H2", path)):
            if not ignore_content:
                content_accumulator.append(text)
            continue

        header_match = re.match(r"//Document/(H1|H2)", path)
        if header_match:
            header_level = header_match.group(1)
            if header_level == "H2" and any(sub in text for sub in ignore_subsections):
                ignore_content = True
                continue
            else:
                ignore_content = False

            if content_accumulator:
                if(len(content_accumulator)==1):
                    print('Empty')
                else:    
                    save_file(content_accumulator, page_counter)
                    page_counter += 1
                    content_accumulator = []

        if not ignore_content:
            content_accumulator.append(text)

    if content_accumulator and not ignore_content:
        save_file(content_accumulator, page_counter)






```

```python
import os
import  shutil
#Conversion
output_dir='output/'
for folder_name, _, files in os.walk(output_dir):
        # Iterate over each file in the current folder
         
         for file_name in files:
            # Check if the file is a JSON file
            
            if file_name.endswith('.json'):
                #print(folder_name)
                # Construct the full path to the JSON file
                file_path = os.path.join(folder_name, file_name)
                # Load the JSON data from the file
                if os.path.exists(folder_name+'/'+'OUT'):
                    print(f"Directory '{folder_name+'/'+'OUT'}' already exists. Deleting...")
                    try:
                        shutil.rmtree(folder_name+'/'+'OUT')
                        print(f"Directory '{folder_name+'/'+'OUT'}' deleted successfully.")
                    except OSError as e:
                        print(f"Error: {e.strerror}")
                os.makedirs(folder_name+'/'+'OUT')
                process_json_and_save_files(file_path, folder_name+'/'+'OUT')
                    
```

```python
import pandas as pd
import numpy as np
```

```python
def generate_table_text(file, tabletxtfn):
    excel_file_path = file
    df = pd.read_excel(excel_file_path, header=None)
    for str_col in df.select_dtypes(include=['object']).columns:
        df[str_col] = df[str_col].astype(str).replace({r"_x([0-9a-fA-F]{4})_": ""}, regex=True)
        df = df.replace(['nan', 'None'], '')
    f = open(tabletxtfn, "w")
    for i in range(df.shape[0]):
        j = 0
        txt = ""
        while j <= (df.shape[1]-1):
            
            txt1 = df.iloc[i, j]
            if not isinstance(txt1, str):
                txt1 = str(txt1)
            if txt1=="":
                j = j+1
                continue
            txt1 = ' '.join(txt1.split())
            txt = txt + " " + txt1
            j = j+1
        f.write (txt + "\n")
    f.close()
```

```python
#Table to text
import os
output_dir='output/'
for root, dirs, files in os.walk(output_dir):
        for file in files:
            if file.endswith('.xlsx'):
                # Read Excel file
                #print(root + ' ' +file)
                excel_file_path = os.path.join(root, file)
                txt_file_path = os.path.join(root, file.replace(".xlsx", ".txt"))
                generate_table_text(excel_file_path, txt_file_path)
```

```python
#Putting back the table
def get_table_text(i,root_dir):
    content=''
    for foldername, _, filenames in os.walk(root_dir):
        for filename in filenames:
            filepath = os.path.join(foldername, filename)
            if filename.startswith(f'fileoutpart{i}') and filename.endswith('.txt'):
                with open(filepath, 'r+') as file:
                  content=file.read()
    return content   



def process_directory(root_dir):
    index = 0
    newdir=root_dir+'/modified'
    #print('newdir',newdir)
    if os.path.exists(newdir):
        print(f"Directory '{newdir}' already exists. Deleting...")
        try:
            shutil.rmtree(newdir)
            print(f"Directory '{newdir}' deleted successfully.")
        except OSError as e:
            print(f"Error: {e.strerror}")
    os.makedirs(newdir)
    for foldername, _, filenames in os.walk(root_dir):
        #print('foldername:',foldername)
        if(foldername==newdir):
            continue
        for filename in filenames:
            if filename.endswith('.txt'):
                if filename.startswith('fileoutpart') and filename.endswith('.txt'):
                    continue
                
                        
                input_filepath = os.path.join(foldername, filename)
                output_filepath = os.path.join(newdir, filename)  # Output file path
                with open(input_filepath, 'r') as input_file:
                    content = input_file.read()  # Read the entire content of the input file
                   
                    while '###TABLE###' in content:
                        table_content = get_table_text(index, root_dir)
                        if filename.startswith('section_0') and filename.endswith('.txt'):
                            print(input_filepath,table_content)
                        if table_content:
                            content = content.replace('###TABLE###', table_content, 1)  # Replace only the first occurrence
                            index += 1  # Increment index
                        else:
                            break  # Break if no table content found
                with open(output_filepath, 'w') as output_file:  # Write mode to overwrite existing content
                    output_file.write(content)  # Write the modified content to the output file
                        
```

```python
#Give the folders where all the files of extracted pdfs are present
base='output/'
files=os.listdir(base)

for file in files:
    process_directory(base+file)
    
    
```
